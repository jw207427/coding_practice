```python
class node:
    def __init__(self, value):
        self.value = value
        self.next = None
        
    def insert(self, value):
        if not self.next:
            self.next = node(value)
        else:
            self.next.insert(value)
            
    def display(self):
        print(self.value)
        if self.next:
            self.next.display()
            
def Merge_sort(h):
    if not h:
        return
    if not h.next:
        return h
    
    midd = find_middle(h)
    right = midd.next
    
    midd.next = None
#     h.display()
#     print('==========')
#     right.display()
#     print('==========')
    
    left = Merge_sort(h)
    right = Merge_sort(right)
#     h.display()
#     print('==========')
#     right.display()
    
    return merge_nodes(left, right)


def merge_nodes(left, right):
    if not left:
        return right
    if not right:
#         print(left.value)
        return left
    
    temp = None
#     left.display()
#     print('==========')
#     right.display()
#     print('==========')
    if left.value < right.value:
        temp = left
        #temp.display()
        temp.next = merge_nodes(left.next, right)
        
    else:
        temp = right
        temp.next = merge_nodes(left, right.next)
    
    return temp
    
    
    
    
def find_middle(h):
    runner = h.next
    slow = h
    
    while runner:
        runner = runner.next
        if not runner:
            break
        runner = runner.next
        slow = slow.next
    
    return slow
    
            
ll = node(3)
A = [4, 2,1, 8, -1]
for a in A:
    ll.insert(a)
        
node = Merge_sort(ll)
node.display()
```
